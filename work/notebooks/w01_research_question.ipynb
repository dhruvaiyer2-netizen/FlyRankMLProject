## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

My work produces a ranked review queue: pages that rank well but earn fewer clicks or less engagement than similar pages at the same position, sorted by the size of that gap. A content/SEO reviewer with limited weekly capacity acts on it — they open the top of the list, check the reason code, and decide whether to rewrite a title/meta description, improve the on-page content, or leave it alone.
A wrong call in either direction costs something different. A false positive — flagging a page that's actually fine (just seasonal, or a small-volume fluke) — wastes reviewer time and risks an unnecessary edit to something that wasn't broken. A false negative — missing a page that's genuinely under-earning clicks — means real traffic keeps leaking silently, with no one prompted to look. Since review capacity is limited, the queue's job is to make sure that capacity is spent on the highest-probability real opportunities first, not to catch every single case.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

The numbers show something interesting: the median ctrs for both ranking 1-3 and 4-10 are the same, while mean ctrs are different. The expected steep drop-off in ctrs is visible in the mean but very weak in the median, suggesting a subset of high performing pages in the top tiers, rather than a uniform advantage for all well positioned pages.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import os, sys, subprocess

IN_COLAB = "google.colab" in sys.modules
REPO_URL = "https://github.com/flyrank-bih/flyrank-ml-internship-starter"
REPO_DIR = "flyrank-ml-internship-starter"

if IN_COLAB:
    if not os.path.isdir(REPO_DIR):
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)
    os.chdir(REPO_DIR)
    subprocess.run([sys.executable, "-m", "pip", "install", "-q", "-r", "requirements.txt"], check=True)
elif os.path.basename(os.getcwd()) == "notebooks":
    os.chdir("..")
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Apply the same minimum-volume filter the starter pipeline uses
visible = df[(df["impressions_90d"] > 0) & (df["content_age_days"] >= 90)]

# 1. How many visible pages exist at all — the scale of the review pool
n_visible = len(visible)

# 2. How many are page-1-ish AND meaningfully visible (candidate pool before any CTR check)
strong_position = visible[(visible["avg_position"] > 0) & (visible["avg_position"] <= 20)]
n_strong_position = len(strong_position)

print(f"Visible pages (min-volume filter): {n_visible}")
print(f"Visible pages ranking 1-20:        {n_strong_position}")
tiers = [(1, 3), (4, 10), (11, 20)]

volume_filtered = df[df["impressions_90d"] >= 500]

sum_below_median = 0

for lo, hi in tiers:
    tier_df = volume_filtered[(volume_filtered["avg_position"] >= lo) & (volume_filtered["avg_position"] <= hi)]
    tier_median = tier_df["ctr"].median()
    below_median = tier_df[tier_df["ctr"] < tier_df["ctr"].median() * 0.5]  # meaningfully below, not just below
    print(f"Positions {lo}-{hi}: n={len(tier_df)}, median CTR={tier_median:.3f}, meaningfully-below={len(below_median)}")
    sum_below_median += len(below_median)

print(f"Total meaningfully-below: {sum_below_median}")
```
